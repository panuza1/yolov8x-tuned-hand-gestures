```python
import torch
from functools import partial

# บังคับให้ torch.load ใช้ weights_only=False default 
torch.load = partial(torch.load, weights_only=False)

from ultralyticsplus import YOLO, render_result

model = YOLO('lewiswatson/yolov8x-tuned-hand-gestures')   

# test 
image = 'https://thumbs.dreamstime.com/b/young-man-showing-peace-sign-isolated-over-yellow-background-133210696.jpg' 
results = model.predict(image) 

render = render_result(model=model, image=image, result=results[0])
render.show() 

# watch data 
for result in results:
    boxes = result.boxes
    for box in boxes:
        label = model.names[int(box.cls)] 
        conf = float(box.cls)
        print(f"เจอท่าทาง{label} (confidence {conf:.2f})")
```
